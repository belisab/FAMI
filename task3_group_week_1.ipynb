```python
# This file is for the task of extracting news headlines.
import requests
from bs4 import BeautifulSoup
import pandas as pd
import html
import json
```

```python
# Website url
url = "https://www.helsinki.fi/en/news/news-and-press-releases"

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
print(soup)
```

The goal is to collect news information in the overview: (title, description, and link). After searching for the title of the first news item using Ctrl + F in the soup output, we identified that the relevant block was called hy-general-list and that the data were contained in the data-items attribute.

```python
news_items = []

block = soup.find("hy-general-list")
raw_data = block.get("data-items")
decoded = html.unescape(raw_data) # convert HTML-encoded text into readable text
print(decoded)
items = json.loads(decoded) # convert the JSON-formatted string into python objects (list of dictionaries)
print(items)

for item in items:
    title = item.get("title")
    description = item.get("description")
    url = item.get("url")

    news_items.append({
        "title": title,
        "description": description,
        "url": url
    })
for news_items in news_items[0:5]:
    print('TITOLO', news_items.get("title"))
    print('DESCRIZIONE', news_items.get("description"))
    print('URL', news_items.get("url"))
```
